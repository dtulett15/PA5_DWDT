- Problem 1: Hardcode a 20 item dictionary that translates English to French when the user inputs one of the key words
- Problem 2: Print a table that shows the data for a D&D player class that the user inputs

```python
#Problem 1
eng2frn = {
    "school": "école",
    "book": "livre",
    "pen": "stylo",
    "apple": "pomme",
    "potato": "pomme de terre",
    "mushroom": "champignon",
    "man": "homme",
    "woman": "femme",
    "girl": "fille",
    "boy": "garçon",
    "dog": "chien",
    "cat": "chat",
    "bird": "oiseau",
    "night": "nuit",
    "snow": "neige",
    "water": "eau",
    "hand": "main",
    "foot": "pied",
    "arm": "bras",
    "leg": "jambe",
}


x = input("Enter a word to translate from English to French (enter 0 to stop): ")

while x != "0":
    if x in eng2frn:
        print(f'The French word for {x} is {eng2frn.get(x).upper()}')
    else:
        print(f'The French word for {x.upper()} is not in this dictionary.') 
    x = input("Enter a word to translate from English to French (enter 0 to stop): ")
    
```

```python
#Problem 2
from prettytable import PrettyTable
pt = PrettyTable()

player_classes = {
    "Artificer":[("Hit Die","d8"), 
                 ("Primary Ability", "Intelligence"),
                 ("Saves", "Constitution & Intelligence"),
                 "Masters of unlocking magic in everyday objects, artificers are supreme inventors."
                ],
    "Blood Hunter":[("Hit Die","d10"), 
                 ("Primary Ability", "Strength or Dexterity, & Intelligence"),
                 ("Saves", "Dexterity & Intelligence"),
                 "A fanatical slayer that embraces dark knowledge to destroy evil."
                ],
    "Druid":[("Hit Die","d8"), 
                 ("Primary Ability", "Wisdom"),
                 ("Saves", "Intelligence & Wisdom"),
                 "A priest of the Old Faith, wielding the powers of nature and adopting animal forms."
                ],
    "Paladin":[("Hit Die","d10"), 
                 ("Primary Ability", "Strength & Charisma"),
                 ("Saves", "Wisdom & Charisma"),
                 "A holy warrior bound to a sacred oath."
                ],
    "Rogue":[("Hit Die","d8"), 
                 ("Primary Ability", "Dexterity"),
                 ("Saves", "Dexterity & Intelligence"),
                 "A scoundrel who uses stealth and trickery to overcome obstacles and enemies."
                ],
    "Wizard":[("Hit Die","d6"), 
                 ("Primary Ability", "Intelligence"),
                 ("Saves", "Intelligence & Wisdom"),
                 "A scholarly magic-user capable of manipulating the structures of reality."
                ],
}


def printTable(ddclass):
    """Given a key from player_class dict, prints a table of
    data for that class."""
    title1, entry1 = player_classes.get(ddclass)[0]
    title2, entry2 = player_classes.get(ddclass)[1]
    title3, entry3 = player_classes.get(ddclass)[2]
    
    pt.field_names = [title1, title2, title3]
    pt.add_row([entry1, entry2, entry3])
    
    print(pt)
    
x = input("Enter the class of player you would like to look up. Enter zero (0) to quit. ")
while x != "0":
    if x in player_classes:
        print()
        print(f'Description:  {player_classes.get(x)[3]}')
        printTable(x)
        print()
        pt.clear_rows()
    else:
        print("I have no record of this player class!")
        print()

    x = input("Enter the class of player you would like to look up. Enter zero (0) to quit. ")
```
